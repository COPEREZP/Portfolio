```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import os
import time

# ---------------- CONFIG ----------------
chrome_options = Options()
chrome_options.add_argument("--disable-gpu")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")

driver = webdriver.Chrome(
    service=Service(ChromeDriverManager().install()),
    options=chrome_options
)

wait = WebDriverWait(driver, 10)

productos = []

# ---------------- RUTA DINÁMICA ----------------
def obtener_ruta_descargas():
    return os.path.join(os.path.expanduser("~"), "Downloads", "KILLAH4.xlsx")

# ---------------- SCRAPER ----------------
def extraer_productos(url):
    driver.get(url)

    try:
        # Esperar a que carguen productos
        wait.until(EC.presence_of_element_located(
            (By.CLASS_NAME, "vtex-product-summary-2-x-container")
        ))
    except:
        print("No cargaron productos → STOP")
        return False

    product_elements = driver.find_elements(By.CLASS_NAME, 'vtex-product-summary-2-x-container')

    # 🔴 CORTE REAL
    if len(product_elements) == 0:
        print("No hay productos → STOP")
        return False

    print(f"Productos encontrados: {len(product_elements)}")

    for product in product_elements:
        try:
            # Nombre
            try:
                nombre = product.find_element(
                    By.CSS_SELECTOR,
                    "span.vtex-product-summary-2-x-productBrand"
                ).text
            except:
                nombre = ""

            # Precio
            try:
                precio = product.find_element(
                    By.CSS_SELECTOR,
                    "span.vtex-product-price-1-x-sellingPrice"
                ).text
            except:
                precio = ""

            # Link
            try:
                link = product.find_element(By.TAG_NAME, "a").get_attribute("href")
            except:
                link = ""

            # Imagen
            try:
                imagen = product.find_element(By.TAG_NAME, "img").get_attribute("src")
            except:
                imagen = ""

            productos.append({
                "Nombre": nombre,
                "Precio": precio,
                "URL": link,
                "Imagen": imagen
            })

        except Exception as e:
            print(f"Error producto: {e}")

    return True

# ---------------- LOOP ----------------
for i in range(1, 100):  # no importa si pones 100, se corta solo
    print(f"\nProcesando página {i}")
    url = f'https://www.colloky.cl/accesorios?page={i}'

    continuar = extraer_productos(url)

    if not continuar:
        break

    time.sleep(2)  # pequeño respiro entre páginas

# ---------------- GUARDAR ----------------
df = pd.DataFrame(productos)

if df.empty:
    print("⚠️ No se extrajo ningún producto → revisa selectores o bloqueo web")
else:
    try:
        # Tu ruta original (si existe)
        output_path = r'C:\Users\const\Downloads\KILLAH4.xlsx'
        df.to_excel(output_path, index=False)
    except:
        # Fallback automático
        output_path = obtener_ruta_descargas()
        df.to_excel(output_path, index=False)

    print(f"\n✅ Excel guardado en: {output_path}")
    print(f"Total productos: {len(df)}")

# ---------------- CERRAR ----------------
driver.quit()
```
