```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import time
import pandas as pd
import os

# ---------------- RUTA SEGURA ----------------
def obtener_ruta_excel():
    home = os.path.expanduser("~")

    rutas = [
        os.path.join(home, "Downloads"),
        os.path.join(home, "Descargas"),
        os.path.join(home, "Desktop"),
        os.path.join(home, "Escritorio"),
    ]

    for r in rutas:
        try:
            os.makedirs(r, exist_ok=True)
            return os.path.join(r, "MICHAEL2.xlsx")
        except:
            continue

    return os.path.join(os.getcwd(), "MICHAEL2.xlsx")

# ---------------- SCRAPER ----------------
def scrape_products_multiple_pages():

    # CONFIG SELENIUM
    chrome_options = Options()
    chrome_options.add_argument("--headless=new")  # ✅ fix moderno
    chrome_options.add_argument("--disable-gpu")
    chrome_options.add_argument("--no-sandbox")

    driver = webdriver.Chrome(
        service=Service(ChromeDriverManager().install()),
        options=chrome_options
    )

    output_path = obtener_ruta_excel()
    print(f"\n📁 SE GUARDARÁ EN: {output_path}\n")

    base_url = "https://www.diadora.com/es/cl/ni%C3%B1os-zapatillas/?srule=sorting-in-season&start={}"

    page_number = 0
    product_data = []

    while True:
        url = base_url.format(page_number * 24)
        print(f"\nPágina {page_number + 1}")

        driver.get(url)
        time.sleep(5)

        productos = driver.find_elements(By.CSS_SELECTOR, "div.card-tile")

        # 🔴 CORTE REAL
        if not productos:
            print("No hay más productos → STOP")
            break

        print(f"Productos encontrados: {len(productos)}")

        for producto in productos:
            try:
                nombre = producto.find_element(By.CSS_SELECTOR, "span.product_title_name").text

                precio = producto.find_element(By.CSS_SELECTOR, "span.price__basic").text.strip()

                try:
                    imagen = producto.find_element(By.CSS_SELECTOR, "img.img-fluid").get_attribute("data-src")
                except:
                    imagen = ''

                link = producto.find_element(By.CSS_SELECTOR, "a.itemProduct__link").get_attribute("href")

                product_data.append({
                    "Nombre": nombre,
                    "Precio": precio,
                    "Imagen": imagen,
                    "URL": link
                })

                print(f"✔ {nombre}")

            except Exception as e:
                print(f"Error producto: {e}")

        page_number += 1
        time.sleep(2)

    # ---------------- GUARDADO FINAL (CLAVE) ----------------
    print("\n💾 Guardando Excel...")

    if len(product_data) == 0:
        print("❌ No se extrajo nada → revisa la web o selectores")
    else:
        try:
            df = pd.DataFrame(product_data)
            df.to_excel(output_path, index=False)
            print(f"✅ GUARDADO EXITOSO en: {output_path}")
            print(f"Total productos: {len(df)}")
        except Exception as e:
            print(f"❌ ERROR GUARDANDO: {e}")

    driver.quit()

# RUN
scrape_products_multiple_pages()
```
